```python
!pip install matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 준비
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 1. Figure 객체 생성 (전체 캔버스)
fig = plt.figure(figsize=(10, 5))

# 2. Figure 안에 Axes 객체(서브플롯) 추가
# 1행 2열의 첫 번째 위치에 ax1 생성
ax1 = fig.add_subplot(1, 2, 1)
# 1행 2열의 두 번째 위치에 ax2 생성
ax2 = fig.add_subplot(1, 2, 2)

# 3. 각 Axes 객체에 데이터 플로팅
ax1.plot(x, y1, color='blue', label='sin(x)')
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()

ax2.plot(x, y2, color='red', linestyle='--', label='cos(x)')
ax2.set_title('Cosine Wave')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# Figure 전체 제목 설정
fig.suptitle('Sine and Cosine Waves', fontsize=16)

# 레이아웃 자동 조정
plt.tight_layout()

# 화면에 표시
plt.show()
```

```python
# 2행 2열로 서브 플롯 만들기
fig, axes = plt.subplots(2, 2)

# 서브 플롯에 그래프 그리기
axes[0, 0].plot([1, 2, 3], [4, 5, 6])        # 첫 번째 서브 플롯 (Line plot)
axes[0, 1].bar([1, 2, 3], [4, 5, 6])         # 두 번째 서브 플롯 (Bar chart)
axes[1, 0].scatter([1, 2, 3], [4, 5, 6])     # 세 번째 서브 플롯 (Scatter plot)
axes[1, 1].hist([1, 2, 3, 4, 5, 6], bins=3)  # 네 번째 서브 플롯 (Histogram)

# 그래프 간의 간격을 자동으로 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
!pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# iris 데이터셋 로드
iris = sns.load_dataset('iris')

# 품종(species)에 따라 다른 색으로 pairplot 그리기
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
# iris 데이터셋 로드
iris = sns.load_dataset('iris')

# iris 데이터의 상관계수 계산
correlation_matrix = iris.drop(columns='species').corr()

# 히트맵으로 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

```python
# ▶︎ 1. 내장 데이터셋 불러오기
tips = sns.load_dataset('tips')
print(tips.head())  # 데이터 확인

# ▶︎ 2. 히스토그램: 'total_bill'의 분포 확인
plt.figure(figsize=(6, 4))
sns.histplot(data=tips, x='total_bill', bins=10)
plt.title("Histogram of Total Bill")
plt.show()

# ▶︎ 3. 박스플롯: 요일별 'total_bill'의 분포 확인
plt.figure(figsize=(6, 4))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

```python
# 요일(day)별 평균 total_bill 시각화
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Barplot of Total Bill by Day")
plt.show()

# total_bill vs tip, 시간대(time)별 색 구분
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
# 스타일 설정 ('darkgrid', 'whitegrid', 'dark', 'white', 'ticks' 중 선택 가능)
sns.set_style('whitegrid')

# 히스토그램 그리기
sns.histplot(data=tips, x='total_bill')
plt.title("Styled Histogram")
plt.show()

# 막대그래프에 색상 팔레트 적용
sns.barplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title("Barplot with Custom Palette")
plt.show()
```

```python
# 박스플롯에 hue='sex' 적용하여 성별에 따라 색상 분리
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title("Boxplot with Hue")
plt.show()
```

```python
# 다중 서브플롯 예제

# 예시 데이터 로드
tips = sns.load_dataset("tips")

# 2행 2열로 서브플롯 생성
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 첫 번째 서브플롯: 산점도
sns.scatterplot(x="total_bill", y="tip", data=tips, ax=axes[0, 0])

# 두 번째 서브플롯: 박스플롯
sns.boxplot(x="day", y="total_bill", data=tips, ax=axes[0, 1])

# 세 번째 서브플롯: 히트맵
# 숫자형 데이터만 선택하여 상관 행렬 계산
corr = tips.select_dtypes(include=["float64", "int64"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[1, 0])

# 네 번째 서브플롯: 막대 그래프
sns.barplot(x="day", y="total_bill", data=tips, ax=axes[1, 1])

# 그래프 간의 간격 자동 조정
plt.tight_layout()

```

```python
# 데이터셋 불러오기
tips = sns.load_dataset("tips")

# Seaborn 스타일 설정
sns.set(style="whitegrid")

# Seaborn으로 countplot 생성 (요일 순서 지정)
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='day', data=tips, order=['Thur', 'Fri', 'Sat', 'Sun'])

# Matplotlib로 세부 설정
ax.set_title('Count of Diners by Day', fontsize=16)
ax.set_xlabel('Day of the Week', fontsize=12)
ax.set_ylabel('Count', fontsize=12)

# 각 막대 위에 수치 표시
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}',                     # 텍스트 내용
                (p.get_x() + p.get_width() / 2, height),  # 텍스트 위치
                ha='center', va='top',           # 수평, 수직 정렬
                color='black', size=12)          # 색상과 크기

# 그래프 출력
plt.show()
```
