```python
import random
import time
import math
from multiprocessing import Pool, cpu_count

# 소수 판별 함수
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.isqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# 단일 프로세스 소수 세기
def count_primes_single(numbers):
    start = time.perf_counter()
    count = sum(1 for num in numbers if is_prime(num))
    end = time.perf_counter()
    return count, end - start

# 멀티 프로세스 소수 세기
def count_primes_multi(numbers):
    start = time.perf_counter()
    with Pool(processes=cpu_count()) as pool:
        results = pool.map(is_prime, numbers)
    count = sum(results)
    end = time.perf_counter()
    return count, end - start

if __name__ == "__main__":
    N = 10_000_000 
    numbers = [random.randint(1, 10_000_000) for _ in range(N)]

    # 단일 프로세스
    sp_count, sp_time = count_primes_single(numbers)
    print(f"[단일] 소수 개수: {sp_count}, 시간: {sp_time:.2f}초")

    # 멀티 프로세스
    mp_count, mp_time = count_primes_multi(numbers)
    print(f"[멀티] 소수 개수: {mp_count}, 시간: {mp_time:.2f}초")

```
