```python
# 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. 샘플 데이터 생성 (클러스터링용 가상 데이터)
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 2. KMeans 모델 생성 및 학습
# n_clusters=4: 4개의 클러스터로 묶음
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X)  # 학습 + 예측 동시에 수행

# 3. 클러스터 중심 좌표 확인
centers = kmeans.cluster_centers_
print("클러스터 중심점 좌표:")
print(centers)

# 4. 시각화
plt.figure(figsize=(8, 6))

# 데이터 포인트를 클러스터별로 색상 구분하여 표시
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', alpha=0.7)

# 클러스터 중심점 시각화
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, marker='X', label='Centroids')

plt.title('K-Means Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 5. Inertia 값 확인 (클러스터 응집도 평가 지표)
print("Inertia (총 제곱 거리합):", kmeans.inertia_)
```
