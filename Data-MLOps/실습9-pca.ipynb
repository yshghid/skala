```python
!pip install scikit-learn

# ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (100Í∞ú ÏÉòÌîå, 10Í∞ú ÌäπÏÑ±)
np.random.seed(42)
X = np.random.rand(100, 10)  # 0~1 ÏÇ¨Ïù¥ ÎûúÎç§Í∞í

# 2. Îç∞Ïù¥ÌÑ∞ Ïä§ÏºÄÏùºÎßÅ (ÌëúÏ§ÄÌôî)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA Î™®Îç∏ ÏÉùÏÑ± Î∞è ÌïôÏäµ (2Ï∞®ÏõêÏúºÎ°ú Ï∂ïÏÜå)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 4. Í≤∞Í≥º Ï∂úÎ†•
print("üîπ Original data shape:", X_scaled.shape)
print("üîπ PCA transformed data shape:", X_pca.shape)
print("üîπ Explained variance ratio (Í∞Å Ï£ºÏÑ±Î∂ÑÏùò Í∏∞Ïó¨ÎèÑ):", pca.explained_variance_ratio_)
print("üîπ Sum of explained variance ratio (Ï¥ù ÏÑ§Î™ÖÎêú Î∂ÑÏÇ∞ ÎπÑÏú®):", np.sum(pca.explained_variance_ratio_))

# 5. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.grid(True)
plt.show()
```
